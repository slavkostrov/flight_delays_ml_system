```python
import boto3
session = boto3.session.Session()
s3 = session.client(
    service_name='s3',
    endpoint_url='https://storage.yandexcloud.net'
)
```

```python
from tqdm.auto import tqdm

# hold only useful columns
columns = ['FL_DATE',
 'OP_CARRIER',
 'OP_CARRIER_FL_NUM',
 'ORIGIN',
 'DEST',
 'CRS_DEP_TIME',
 'DEP_TIME',
 'DEP_DELAY',
 'TAXI_OUT',
 'WHEELS_OFF',
 'WHEELS_ON',
 'TAXI_IN',
 'CRS_ARR_TIME',
 'ARR_TIME',
 'ARR_DELAY',
 'CANCELLED',
 'DIVERTED',
 'CRS_ELAPSED_TIME',
 'ACTUAL_ELAPSED_TIME',
 'AIR_TIME',
 'DISTANCE']
for file_name in tqdm(sorted(os.listdir(DATA_PATH))):
    file_path = f"{DATA_PATH}/{file_name}"
    data = pd.read_csv(file_path)
    
    # drop rows with NA target
    data.dropna(inplace=True, axis=0, subset=["DEP_DELAY"])
    
    # add 5 years to simulate nowadays data
    data["FL_DATE"] = pd.to_datetime(data["FL_DATE"]) + pd.to_timedelta(365 * 5, unit="days")
    data = data[data["FL_DATE"].dt.year == data["FL_DATE"].dt.year.mode()[0]]
    year = pd.to_datetime(data["FL_DATE"]).dt.year.mode()[0]
    
    # print some info
    print(file_name, data.shape, list(data.columns))

    # select only useful columns
    data = data[columns]
    
    # form new name of file
    new_file = f"{year}.parquet"
    
    # writing temp parquet file 
    data.to_parquet("data.parquet", index=False)
    
    print("uploading data to s3...")
    s3.upload_file("data.parquet", "slava-otus", f"final_project/features/{new_file}")
    forDeletion = [{'Key': f"final_project/features/{file_name}"}]
    response = s3.delete_objects(Bucket='slava-otus', Delete={'Objects': forDeletion})
```
